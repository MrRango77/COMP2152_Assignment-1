```python
# Step b: Create 4 variables
gym_member = "Alex Alliton"     # str
preferred_weight_kg = 20.5      # float
highest_reps = 25              # int
membership_active = True        # bool
```

```python
# Step c: Create a dictionary named workout_stats
# Data type: dict where values are tuples of three ints (yoga, running, weightlifting)
workout_stats = {
    "Alex": (30, 45, 20),
    "Jamie": (40, 35, 50),
    "Taylor": (25, 40, 30),
}
```

```python
# Step d: Calculate total workout minutes using a loop and add to dictionary
friends = list(workout_stats.keys())  # store friend names before adding *_Total keys

for friend in friends:
    yoga_min, running_min, lifting_min = workout_stats[friend]
    total_minutes = yoga_min + running_min + lifting_min
    workout_stats[f"{friend}_Total"] = total_minutes

print("Workout stats with totals added:")
print(workout_stats)
```

```python
# Step e: Create a 2D nested list called workout_list
# 2D list (list of lists) where each row is a friend and each column is an activity (yoga, running, weightlifting)
workout_list = []

for friend in friends:
    yoga_min, running_min, lifting_min = workout_stats[friend]
    workout_list.append([yoga_min, running_min, lifting_min])

print("workout_list (rows=friends, cols=activities):")
print(workout_list)
```

```python
# Step f: Slice the workout_list
# Extract and print the minutes for yoga and running for all friends (cols 0 and 1)
yoga_running_all = [row[0:2] for row in workout_list]
print("Yoga and running minutes for all friends:")
print(yoga_running_all)

# Extract and print the minutes for weightlifting for the last two friends (last 2 rows, col 2)
weightlifting_last_two = [row[2] for row in workout_list[-2:]]
print("Weightlifting minutes for the last two friends:")
print(weightlifting_last_two)
```

```python
# Step g: Check if any friend's total >= 120
for friend in friends:
    total_key = f"{friend}_Total"
    if workout_stats[total_key] >= 120:
        print(f"Great job staying active, {friend}!")
```

```python
# Step h: User input to look up a friend
name_input = input("Enter a friend's name to look up: ").strip()

# Make lookup case-insensitive while still using original names
name_map = {n.lower(): n for n in friends}
lookup_key = name_input.lower()

if lookup_key in name_map:
    friend = name_map[lookup_key]
    yoga_min, running_min, lifting_min = workout_stats[friend]
    total_minutes = workout_stats[f"{friend}_Total"]

    print(f"Workout minutes for {friend}:")
    print("Yoga:", yoga_min)
    print("Running:", running_min)
    print("Weightlifting:", lifting_min)
    print("Total:", total_minutes)
else:
    print(f"Friend {name_input} not found in the records.")
```

```python
# Step i: Friend with highest and lowest total workout minutes
highest_friend = friends[0]
lowest_friend = friends[0]

for friend in friends:
    if workout_stats[f"{friend}_Total"] > workout_stats[f"{highest_friend}_Total"]:
        highest_friend = friend
    if workout_stats[f"{friend}_Total"] < workout_stats[f"{lowest_friend}_Total"]:
        lowest_friend = friend

print("Friend with the highest total workout minutes:", highest_friend, "-", workout_stats[f"{highest_friend}_Total"])
print("Friend with the lowest total workout minutes:", lowest_friend, "-", workout_stats[f"{lowest_friend}_Total"])
```
